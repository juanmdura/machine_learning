```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from word2number import w2n
```

```python
df = pd.read_csv('hiring.csv')
df
```

```python
df.experience = df.experience.fillna('zero')
df
```

```python
df.columns = [c.replace(' ', '_') for c in df.columns]
df.columns = [c.replace('(', '_') for c in df.columns]
df.columns = [c.replace(')', '') for c in df.columns]
df.columns = [c.replace('$', '') for c in df.columns]
df
```

```python
import math
median_test = math.floor(df.test_score_out_of_10.median())
median_test
```

```python
df.test_score_out_of_10 = df.test_score_out_of_10.fillna(median_test)
df
```

```python
# w2n.word_to_num
df.experience = df.experience.apply(w2n.word_to_num)
df.experience
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience', 'test_score_out_of_10','interview_score_out_of_10']].values, df.salary_)
```

```python
reg.coef_
```

```python
reg.intercept_
```
